# 라이브러리 임포트

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
```

# 로드 및 테스트

- Ko-SRoBERTa 모델 로드
- 예제 문장을 임베딩 벡터로 변환 및 출력

```python
encoder = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
embeddings = encoder.encode(sentences)

print(embeddings)
```

# 데이터 로드

- CSV 파일을 불러와 데이터프레임으로 변환
- 데이터 확인(첫 5개 행 출력)

```python
df = pd.read_csv('dataset.csv')

df.head()
```

# 필요하지 않은 열 제거

- 데이터셋에서 사용하지 않는 열 제거

```python
df = df.drop(columns=['Unnamed: 3'])

df.head()
```

# 누락된 값 제거

- 데이터셋에서 결측값 제거
- 데이터 크기 확인

```python
print(len(df))
df = df.dropna()
print(len(df))
```

# 특정 데이터 확인

- 첫번째 행에서 유저란 열 확인

```python
df.loc[0, '유저']
```

# 첫 번째 문장의 임베딩 생성

- 첫번째 문장을 임베딩 벡터로 변환

```python
encoder.encode(df.loc[0, '유저'])
```

# 임베딩 열 추가

- 각 문장을 임베딩하고 새로운 열로 추가

```python
df['embedding'] = pd.Series([[]] * len(df))
df['embedding'] = df['유저'].map(lambda x: list(encoder.encode(x)))
df.head()
```

# 입력 문장 임베딩

- 입력 문장을 임베딩 벡터로 변환

```python
text = '요즘 남편이 속을 너무 썩여 술도 맨날 마셔'
embedding = encoder.encode(text)
embedding
```

# 유사도 계산

- 유저 입력과 데이터셋 문장 간 코사인 유사도를 계산
- 결과 저장

```python
df['similarity'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())

df.head()
```

# 가장 유사한 문장 찾기

- 가장 높은 유사도를 가진 문장을 찾음
- 해당 답변과 유사도를 출력

```python
answer = df.loc[df['similarity'].idxmax()]

print('구분', answer['구분'])
print('유사한 질문', answer['유저'])
print('챗봇 답변', answer['챗봇'])
print('유사도', answer['similarity'])
```

# 인터페이스 구현 및 실행

```python
import gradio as gr

def greet(user):
    embedding = encoder.encode(user)

    df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())
    answer = df.loc[df['distance'].idxmax()]

    return answer['챗봇']

demo = gr.Interface(fn=greet, inputs="text", outputs="text")

demo.launch(debug=True, share=True)
```
